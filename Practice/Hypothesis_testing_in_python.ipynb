```python
dataset=sns.load_dataset('tips')
```

```python
dataset_table=pd.crosstab(dataset['sex'],dataset['smoker'])
```

```python
dataset_table.values
```

```python
observed_values = dataset_table.values
```

```python
observed_values
```

```python
from scipy import stats
```

```python
stats_test,p,dof,expected_values = stats.chi2_contingency(observed_values)
```

```python
p
```

```python
dataset_table
```

```python
no_of_rows = dataset_table.iloc[0:2,0]
```

```python
no_of_columns = dataset_table.iloc[0,0:2]
```

```python
degoffreedom = (len(no_of_rows)-1)*(len(no_of_columns)-1)
```

```python
degoffreedom
```

```python
list(zip(observed_values,expected_values))
```

```python
import numpy as np
sum(np.array([5,6])-np.array([4,3]))
```

```python
for o,e in zip(observed_values,expected_values):
    print(o,e,)
```

```python
from scipy.stats import chi2
```

```python
chisquare_test=sum([(o-e)**2./e for o,e in zip(observed_values,expected_values)])
```

```python
chi_square_statistic = sum(chisquare_test)
```

```python
chi_square_statistic
```

```python
alpha = 0.05
```

```python
critical_value = chi2.ppf(1-alpha,df= degoffreedom)
```

```python
critical_value
```

```python
if chi_square_statistic>=critical_value:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```

```python
#p-value
p_value=1-chi2.cdf(x=chi_square_statistic,df=degoffreedom)
```

```python
if p_value<=alpha:
    print("Reject H0,There is a relationship between 2 categorical variables")
else:
    print("Retain H0,There is no relationship between 2 categorical variables")
```
