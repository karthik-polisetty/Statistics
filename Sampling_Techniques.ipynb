```python
import numpy as np
import pandas as pd

# generating population data that follows normal distribution
mu = 5
std = 1
size = 1000
pop_df = np.random.normal(mu,std,size)

# function to create random sampling

def random_sampling(df,n):
    sample = np.random.choice(df,replace = False, size = n)
    return sample

random_sampling(pop_df,50)
```

```python
import numpy as np
import pandas as pd

# generating population data that follows normal distribution
mu = 5
std = 1
size = 1000
pop_df = np.random.normal(mu,std,size)

#function for systematic sampling

def systematic_sampling(df,sample_size):
    step = int(size/sample_size)
    id = pd.Series(np.arange(1,len(df)+1))
    df = pd.Series(df)
    main_df = pd.concat([id,df],axis=1)
    main_df.columns = ['id','data']
    main_df.set_index('id')
    sample_indexes = np.arange(0,len(df),step)
    sample = main_df.iloc[sample_indexes]
    
    return sample

systematic_sampling(pop_df,10)

    
    
```

```python
import numpy as np
import pandas as pd

#create a dataframe
price = pd.Series(np.random.uniform(1,5,size = 1000))
id = pd.Series(np.arange(0,len(price),1))
typeofevent = pd.Series(np.random.choice(['type1','type2','type3'],size = len(price)))
click = pd.Series(np.random.choice([0,1],size = len(price)))

df= pd.concat([id,price,typeofevent,click],axis = 1)
df.columns = ['id','price','typeofevent','click']


def clustered_sampling(df,clustersize,clusters_req):
    #clusters_req = no of clusters we want 
    k = int(len(df)/clustersize)
    data = None
    
    for i in range(k):
        sample_k = df.sample(clustersize)
        sample_k["cluster"] = np.repeat(i,len(sample_k))
        df=df.drop(index = sample_k.index)
        
        data = pd.concat([data,sample_k],axis = 0)
    
    req = np.random.randint(0,k,size = clusters_req)
    
    
    sample = data[data.cluster.isin(req)]
    return sample

clustered_sampling(df,100,3)
```

```python
import numpy as np
import pandas as pd

#create a dataframe
price = pd.Series(np.random.uniform(1,5,size = 10000))
id = pd.Series(np.arange(0,len(price),1))
typeofevent = pd.Series(np.random.choice(['type1','type2','type3'],size = len(price)))
click = pd.Series(np.random.choice([0,1],size = len(price)))

df= pd.concat([id,price,typeofevent,click],axis = 1)
df.columns = ['id','price','typeofevent','click']
```

```python
def stratified_sampling(df,sample_size, no_of_clusters_req):
    N=len(df)
    k = int(N/sample_size) #no of clusters in which data should be divided into
    print(N,k)
    
    def weighted_sampling(df,n):
        def get_class_prob(x):
            weight = int(np.rint(n* (len(x[x.click!=1])/(len(df[df.click!=1])))))
            sample_x = x.sample(weight)
            return (sample_x)
    
        weighted_sample = df.groupby('typeofevent').apply(get_class_prob)
        # print(weighted_sample["typeofevent"].value_counts())
        # print(weighted_sample["click"].value_counts())
        return weighted_sample
    
    stratas = None
    for i in range(k):
        sample_k = weighted_sampling(df,sample_size).reset_index(drop=True)
        sample_k["cluster"] = np.repeat(i,len(sample_k))
        
        df.drop(index = sample_k.index)
        stratas = pd.concat([stratas,sample_k],axis = 0)
    
    #print(data)
    selected_strata_clusters= np.random.randint(0,k,size = no_of_clusters_req)
    
    
    stratified_samples = stratas[stratas.cluster.isin(selected_strata_clusters)]
    return stratified_samples

st_sample = stratified_sampling(df,100,4)
    
st_sample
```

```python
def get_startified_sample(df,n,num_clusters_needed):
    N = len(df)
    num_obs_per_cluster = int(N/n)
    K = int(N/num_obs_per_cluster)

    def get_weighted_sample(df,num_obs_per_cluster):
        def get_sample_per_class(x):
            n_x = int(np.rint(num_obs_per_cluster*len(x[x.click !=0])/len(df[df.click !=0])))
            sample_x = x.sample(n_x)
            return(sample_x)
        weighted_sample = df.groupby("typeofevent").apply(get_sample_per_class)
        return(weighted_sample)

    stratas = None
    for k in range(K):
        weighted_sample_k = get_weighted_sample(df,num_obs_per_cluster).reset_index(drop = True)
        weighted_sample_k["cluster"] = np.repeat(k,len(weighted_sample_k))
        stratas = pd.concat([stratas, weighted_sample_k],axis = 0)
        df.drop(index = weighted_sample_k.index)
    selected_strata_clusters = np.random.randint(0,K,size = num_clusters_needed)
    stratified_samples = stratas[stratas.cluster.isin(selected_strata_clusters)]
    return(stratified_samples)

sample = get_startified_sample(df = df,n = 100,num_clusters_needed = 2)
len(sample)
```
