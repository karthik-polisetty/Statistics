```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('kidney_disease.csv')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
def normal(i):
    if i=='normal':
        return 1
    elif i=='abnormal':
        return 0
```

```python
df['rbc'] = df['rbc'].apply(normal)
```

```python
df['rbc'].value_counts()
```

```python
df['pus_cell']=df['pus_cell'].apply(normal)
```

```python
def present(i):
    if i=='present':
        return 1
    elif i=='notpresent':
        return 0
    
```

```python
df['pcclumps']=df['pcclumps'].apply(present)
```

```python
df['bacteria']=df['bacteria'].apply(present)
```

```python
def yes_no(i):
    if i=='yes':
        return 1
    elif i=='no':
        return 0
```

```python
df.columns
```

```python
df['hyptension']=df['hyptension'].apply(yes_no)
```

```python
df['diabetes-mellitus']=df['diabetes-mellitus'].apply(yes_no)
```

```python
df['cor_artery_disease']=df['cor_artery_disease'].apply(yes_no)
```

```python
df['pe']=df['pe'].apply(yes_no)
```

```python
df['anemia']=df['anemia'].apply(yes_no)
```

```python
df.info()
```

```python
def appetite(i):
    if i=='good':
        return 1
    elif i=='poor':
        return 0
```

```python
df['appetite'].value_counts()
```

```python
df['appetite']=df['appetite'].apply(appetite)
```

```python
df
```

```python
def class_fun(i):
    if i == 'ckd':
        return 1
    elif i == 'notckd':
        return 0
```

```python
df['class'] = df['class'].apply(class_fun)
```
